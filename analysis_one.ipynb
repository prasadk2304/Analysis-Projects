Exploratory Data Analysis

On the basis of Gender

```python
ax = sns.countplot(x = 'Gender', data = df)
```

```python
ax = sns.countplot(x = 'Gender', data = df)
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
df.groupby(['Gender'], as_index = False)['Amount'].sum().sort_values(by = 'Amount', ascending = False)
```

```python
sales_gen = df.groupby(['Gender'], as_index = False)['Amount'].sum().sort_values(by = 'Amount', ascending = False)
sns.barplot(x = 'Gender', y = 'Amount', data = sales_gen)
```

On the basis of Age

```python
ax = sns.countplot(x = 'Age Group', data = df, hue = 'Gender')
```

```python
ax = sns.countplot(x = 'Age Group', data = df, hue = 'Gender')
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_gen = df.groupby(['Age Group'], as_index = False)['Amount'].sum().sort_values(by = 'Amount', ascending = False)
sns.barplot(x = 'Age Group', y = 'Amount', data = sales_gen)
```

On the basis of State

```python
sales_state = df.groupby(['State'], as_index = False)['Orders'].sum().sort_values(by = 'Orders', ascending = False).head(10)
sns.set(rc = {'figure.figsize':(20, 5)})
sns.barplot(x = 'State', y = 'Orders', data = sales_state)
```

```python
sales_state = df.groupby(['State'], as_index = False)['Amount'].sum().sort_values(by = 'Amount', ascending = False).head(10)
sns.set(rc = {'figure.figsize':(20, 5)})
sns.barplot(x = 'State', y = 'Amount', data = sales_state)
```

On the basis of Marital Status

```python
ax = sns.countplot(x = 'Marital_Status', data = df)
sns.set(rc = {'figure.figsize':(7, 4)})
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
ax = sns.countplot(x = 'Marital_Status', data = df, hue = 'Gender')
for bars in ax.containers:
    ax.bar_label(bars)
```

On the basis of Occupation

```python
ax = sns.countplot(x = 'Occupation', data = df)
sns.set(rc = {'figure.figsize':(20, 4)})
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_state = df.groupby(['Occupation'], as_index = False)['Amount'].sum().sort_values(by = 'Amount', ascending = False).head(10)
sns.set(rc = {'figure.figsize':(20, 5)})
sns.barplot(x = 'Occupation', y = 'Amount', data = sales_state)
```

On the basis of Product Category

```python
ax = sns.countplot(x = 'Product_Category', data = df)
sns.set(rc = {'figure.figsize':(30, 4)})
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_product = df.groupby(['Product_Category'], as_index = False)['Amount'].sum().sort_values(by = 'Amount', ascending = False).head(10)
sns.set(rc = {'figure.figsize':(20, 5)})
sns.barplot(x = 'Product_Category', y = 'Amount', data = sales_product)
```

```python
sales_pid = df.groupby(['Product_ID'], as_index = False)['Amount'].sum().sort_values(by = 'Amount', ascending = False).head(10)
sns.set(rc = {'figure.figsize':(20, 5)})
sns.barplot(x = 'Product_ID', y = 'Amount', data = sales_pid)
```
